## Data preparation

Here are some details on this dataset:

The file has 34 rows corresponding to the observed combinations of type of ship, year of construction and period of operation. Each row has information on five variables as follows:

* ship type, coded 1-5 for A, B, C, D and E,
* year of construction (1=1960-64, 2=1965-70, 3=1970-74, 4=1975-79),
* period of operation (1=1960-74, 2=1975-79)
* months of service, ranging from 63 to 20,370, and
* damage incidents, ranging from 0 to 53.

Note that there no ships of type E built in 1960-64, and that ships built in 1970-74 could not have operated in 1960-74. These combinations are omitted from the data file.

```python
import pandas as pd
from pandas.io.stata import StataReader
reader = StataReader('ships.dat')
shipDF = reader.data()

shipDF.head()
```

# Challenge 1

Model the damage incident counts with a Poisson Regression

Remember that you will have to create dummy variables for categorical variables, and if you have time bins (like "1960-1964"), you have the option of either treating each bin as a category (and create dummies for each bin). Also remember to add a constant (to model the intercept).

Take a look at the statsmodels summary table, the goodness of fit indicators (Deviance, Pearson's chi square statistic) and the coefficients. Is this a good model?

```python
tmpDF = shipDF.copy()

# Key: categorical field name
# Value: list of abbreviation and post-dummification redundant field
categoricalFields = {'type':['type','E'],
                     'construction':['con','1975-79'],
                     'operation':['op','1975-79']
                     }
# dummify
# Convert all categorical fields in dummy numerical ones
# AND delete one dummified field because it contains redundant information
for (categoricalField, value) in categoricalFields.iteritems():
    (prefix, dummifiedRedundantField) = value
    tmpDF = pd.concat([tmpDF, pd.get_dummies(tmpDF[categoricalField], prefix=prefix)], axis=1)
    if dummifiedRedundantField:
        tmpDF.drop("%s_%s" % (prefix, dummifiedRedundantField), axis=1, inplace=True)

# of course, also delete the categorical fields
for (categoricalField,_) in categoricalFields.iteritems():
    tmpDF.drop(categoricalField, axis=1, inplace=True)
shipDF2 = tmpDF.copy()
shipDF2.head()
```

```python
unpackedShipDF = pd.DataFrame()
unpackedColumns = [c for c in shipDF2.columns] 
#unpackedColumns

for row_i, row in shipDF2.iterrows():
    for duplicated in range(int(row.damage)):
        unpackedShipDF = unpackedShipDF.append(row[unpackedColumns])

print unpackedShipDF.shape
print shipDF2.damage.sum()
        
unpackedShipDF.head()
```

```python
import statsmodels.api as sm

y = unpackedShipDF['damage']
X = unpackedShipDF.copy()
X.drop('damage', axis=1, inplace=True)
sm_X = sm.add_constant(X)
#sm_poisson = sm.Poisson(Y, sm_X)
sm_poisson = sm.GLM(y,X, family=sm.families.Poisson())

# fit the model
sm_poisson_result = sm_poisson.fit()
print sm_poisson_result.summary()
```
